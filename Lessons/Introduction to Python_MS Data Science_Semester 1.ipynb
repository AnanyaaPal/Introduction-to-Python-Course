```python
#inverting a dictionary
def invert_dict(dic_old):
    dic_new={}
    
    for key in dic_old:
        val=dic_old[key]
        if val not in dic_new:
            dic_new[val]=[key]
        else:
            dic_new[val].append(key)
    return dic_new

invert_dict(letter_dict)
```

```python
import requests


def request_map_at(lat, long, satellite=True, zoom=12, size=(400, 400)):
    base = "https://static-maps.yandex.ru/1.x/?"
    params = dict(
        z=zoom,
        size=str(size[0]) + "," + str(size[1]),
        ll=str(long) + "," + str(lat),
        l="sat" if satellite else "map",
        lang="en_US",
    )
    return requests.get(base, params=params, timeout=60)
```

```python
map_response = request_map_at(51.5072, -0.1275)
```

```python
url = map_response.url
print(url)
```

```python
from io import BytesIO  # A library to convert between files and strings

import imageio  # A library to deal with images, https://pypi.org/project/imageio/
import numpy as np  # A library to deal with matrices
```

```python
def is_green(pixels):
    threshold = 1.1
    greener_than_red = pixels[:, :, 1] > threshold * pixels[:, :, 0]
    greener_than_blue = pixels[:, :, 1] > threshold * pixels[:, :, 2]
    green = np.logical_and(greener_than_red, greener_than_blue)
    return green
```

```python
def count_green_in_png(data):
    f = BytesIO(data)
    pixels = imageio.v2.imread(f)  # Get our PNG image as a numpy array
    return np.sum(is_green(pixels))
```

```python
print(count_green_in_png(map_at(*london_location)))
```

```python
def location_sequence(start, end, steps):
    lats = np.linspace(start[0], end[0], steps)  # "Linearly spaced" data
    longs = np.linspace(start[1], end[1], steps)
    return np.vstack([lats, longs]).transpose()
```

```python
location_sequence(geolocate("London"), geolocate("Cambridge"), 5)
```

```python
def show_green_in_png(data):
    pixels = imageio.imread(BytesIO(data))  # Get our PNG image as rows of pixels
    green = is_green(pixels)

    out = green[:, :, np.newaxis] * np.array([0, 1, 0])[np.newaxis, np.newaxis, :]

    buffer = BytesIO()
    imageio.imwrite(buffer, out, format="png")
    return buffer.getvalue()
```

```python
def math_operation(a,b,method='sum'): #the default value of the parameter method is sum
    if method=='sum':
        res=a+b
    if method=='product':
        res=a*b
    return res 

#calling the function with arguments
math_operation(10,20) #we have not specified anything for the paramter method, thus it utilises the default value
math_operation(10,20,'product') #specifying the argument for the parameter method
```

unpacking an argument 

```python
def describe(name,year):
    print(name,'is born in',year)
    return name,year #we use the return key so that we may use the parameters somewhere else in the function as well 

#calling the arguments as a tuple
describe('tome',22)

#passing the arguments as a list
list_args=['tome',1000]
describe(*list_args)

# but the order of the sequence is crucial
describe(*list_args[::-1])

#describe(**dict_args) calls the describe function using 
#the dictionary dict_args as arguments. 
#The double asterisk ** operator is used to unpack the dictionary, 
#and the keys become the parameter names, while the values become the corresponding arguments.
dict_args={'year':1990,'name':'tome'}
describe(**dict_args)  #see the difference between using a single * and a doube **
```

```python
def describe2(*args):
    name,year=args
    print("the name is",name,"the year is",year)
    return args

describe2("ap",2023)  #only one argument can be input into the paramter 'year'
```

```python
#if we enter more than one value in either of the paramters above, it shows an error
#to resolve the error, we may use the * with the paramters as well, to allow multiple assignment of values

def describe2(*args):
    name,*year=args
    print("the name is",name,"the years are",year)
    return args

describe2('ap',1990,2000)
```

```python
#understand the following error: 

def describe2(*args):
    *name,*year=args
    print("the name is",name,"the years are",year)
    return args

describe2('ap','kns',1990,2000)
```
